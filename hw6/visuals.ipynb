```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
import seaborn as sns

sns.set()
```

```python

x = np.linspace(-0.9, 0.1, 100)
dx = np.linspace(0, 1, 100)

alpha = 2
beta = 3

betta = gamma(alpha) * gamma(beta) / gamma(alpha + beta)


y = dx**(alpha -1) * (1-dx)**(beta -1) / betta
y = y / np.sum(y)


fig = plt.figure(figsize=(25, 9))
plt.plot(x, y)
plt.xlim(-0.9, 0.1)
plt.tight_layout()
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel("Coefficient value", fontsize=30)
plt.savefig("report/figures/personal_opinion.pdf")
```
